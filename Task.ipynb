```
public class Employee{
    protected string Name { get; set; }
    protected string Position { get; set; }
    protected float Salary { get; set; }

    public Employee (string name, string position, float salary){
        Name = name;
        Position = position;
        Salary = salary;
    }

    public virtual void PrintInfo(){
        Console.WriteLine($"Имя: {Name}, Должность: {Position}, Зарплата: {Salary}");
    }
}

public class Manager : Employee{
    protected string Department { get; set; }

    public Manager(string name, string position, float salary, string department) : base(name, position, salary){
        Name = name;
        Position = position;
        Salary = salary;
        Department = department;
    }

    public override void PrintInfo(){
        Console.WriteLine($"Имя: {Name}, Отдел: {Department},  Должность: {Position}, Зарплата: {Salary}");
    }
}

public interface IReportable{
    public void GenerateReport(string report);
}

public interface ITeamLeader{
    public void GetTeamMembers();
}

public class ProjectManager : Manager, IReportable, ITeamLeader{
    private string ProjectName { get; set; }
    protected string[] TeamMembers { get; set; }
    protected string Report { get; set; }


    public ProjectManager(string name, string position, float salary, string department, string projectName, string[] teamMembers) : base(name, position, salary, department){
        Name = name;
        Position = position;
        Salary = salary;
        Department = department;
        ProjectName = projectName; 
        TeamMembers = teamMembers;
    }

    

    public override void PrintInfo(){
        Console.WriteLine($"Имя: {Name}, Отдел: {Department},  Должность: {Position}, Зарплата: {Salary}, Название проекта: {ProjectName}");
    }

    public void GenerateReport(string report){
        Report = report;
        Console.WriteLine($"Отчёт: {Report}");
    }

    public void GetTeamMembers(){
        Console.WriteLine("Участники команды:");
        for (int i = 0; i < TeamMembers.Length; i++){
            Console.WriteLine(TeamMembers[i]);
        }
    }
}

Employee employee = new Employee("Кирилл", "C#, JS, Rust, Ruby Senior", 567000);
Manager manager = new Manager("Дмитрий", "React-Senior", 786000, "Web-Site");
ProjectManager projectManager = new ProjectManager("Даниил", "C#-OOP Senior", 234000, "Back-end", "Pokemonchick", ["Александр", "Николай", "Юлия", "Андрей", "Дарья"]);

employee.PrintInfo();

Console.WriteLine();

manager.PrintInfo();

Console.WriteLine();

projectManager.PrintInfo();
projectManager.GenerateReport("Отчёт: разработка C# программы (todo лист)");
projectManager.GetTeamMembers();
```
